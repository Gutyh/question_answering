```python
# exporting the location of the dataset in the ENV path
import os
os.environ['SQUAD_DIR'] = "../datasets/FQuAD"
!echo $SQUAD_DIR
```

```python
# command line parameters and run the script
run_squad_command = """
python run_squad.py
--model_type camembert
--model_name_or_path camembert-base
--do_train
--do_eval
--do_lower_case
--train_file $SQUAD_DIR/train.json
--predict_file $SQUAD_DIR/valid.json
--per_gpu_train_batch_size 12
--learning_rate 3e-5
--num_train_epochs 2.0
--max_seq_length 384
--doc_stride 128
--output_dir models/fr/17032020/"""
run_squad_command = run_squad_command.replace("\n", " ")
!{run_squad_command}
```
